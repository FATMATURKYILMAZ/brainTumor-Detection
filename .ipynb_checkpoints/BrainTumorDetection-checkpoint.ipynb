## üß© 2.1 G√∂rselleri Y√ºkleme ve Etiketleme

```python
img_size = 150
batch_size = 32

# Geli≈ütirilmi≈ü veri artƒ±rma
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_gen = datagen.flow_from_directory(
    data_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)
val_gen = datagen.flow_from_directory(
    data_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)

# Eƒüitim ve doƒürulama veri daƒüƒ±lƒ±mƒ±
train_labels = train_gen.classes
val_labels = val_gen.classes
print("Eƒüitim seti daƒüƒ±lƒ±mƒ±:", np.bincount(train_labels))
print("Doƒürulama seti daƒüƒ±lƒ±mƒ±:", np.bincount(val_labels))

# Sƒ±nƒ±f aƒüƒ±rlƒ±klarƒ±nƒ± hesapla
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_gen.classes),
    y=train_gen.classes
)
class_weights = dict(enumerate(class_weights))
print("Sƒ±nƒ±f aƒüƒ±rlƒ±klarƒ±:", class_weights)
```

## üéØ 2.2 Normalizasyon (0‚Äì255 ‚Üí 0‚Äì1)

```python
def build_model(hp):
    model = Sequential()
    model.add(Conv2D(filters=hp.Int('conv_1_filter', min_value=32, max_value=128, step=16),
                     kernel_size=hp.Choice('conv_1_kernel', values=[3,5]),
                     activation='relu', input_shape=(img_size, img_size, 3)))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(BatchNormalization())

    model.add(Conv2D(filters=hp.Int('conv_2_filter', min_value=64, max_value=256, step=32),
                     kernel_size=hp.Choice('conv_2_kernel', values=[3,5]),
                     activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(BatchNormalization())

    model.add(Flatten())
    model.add(Dense(units=hp.Int('dense_units', min_value=64, max_value=256, step=32), activation='relu'))
    model.add(Dropout(hp.Float('dropout_rate', min_value=0.2, max_value=0.5, step=0.1)))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

## ‚úÇÔ∏è 2.3 Eƒüitim / Test Verisi Ayƒ±rma

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='hyper_tune',
    project_name='brain_mri_tuning'
)

tuner.search(train_gen, validation_data=val_gen, epochs=10, class_weight=class_weights)

best_model = tuner.get_best_models(num_models=1)[0]
```

# ‚úÖ 3. Model Kurulumu (CNN)

### üéØ Ama√ß:
G√∂rsellerden √∂ƒürenerek t√ºm√∂rl√º/t√ºm√∂rs√ºz sƒ±nƒ±flandƒ±rmasƒ± yapmak

CNN (Convolutional Neural Network) ile g√∂rselden √∂znitelik √ßƒ±karmak

## 3.2üß† Model Mimarisi (CNN)

```python
early_stop = EarlyStopping(patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(patience=3, factor=0.5, verbose=1)

history = best_model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=30,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weights
)

# 6. Model Doƒüruluk Grafiƒüi
plt.plot(history.history['accuracy'], label='Eƒüitim')
plt.plot(history.history['val_accuracy'], label='Doƒürulama')
plt.legend()
plt.title('Model Doƒüruluklarƒ±')
plt.grid(True)
plt.show()
```

## 3.3 Model Derleme (Compile)

```python
val_gen.reset()
preds = best_model.predict(val_gen)
y_pred = (preds > 0.5).astype(int).flatten()
y_true = val_gen.classes

print(classification_report(y_true, y_pred, zero_division=1))
sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Tahmin')
plt.ylabel('Ger√ßek')
plt.title('Karƒ±≈üƒ±klƒ±k Matrisi')
plt.show()
```
